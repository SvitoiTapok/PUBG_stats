```python
import matplotlib.pyplot as plt
from lifelines import KaplanMeierFitter
import numpy as np

# 1. Загружаем только необходимые колонки сразу (без создания лишних копий)
analysis_data = death_data[['player_survive_time', 'player_kills']].copy()

# 2. Оптимизация типов данных
analysis_data['player_survive_time'] = analysis_data['player_survive_time'].astype('float32')
analysis_data['player_kills'] = analysis_data['player_kills'].astype('int16')

# 3. Создаем событие (используем меньше памяти)
analysis_data['event'] = (analysis_data['player_survive_time'] <= 1800).astype('bool')

# 4. Группировка с категориальным типом
analysis_data['kill_group'] = pd.cut(
    analysis_data['player_kills'],
    bins=[-1, 0, 2, 4, 7, np.inf],
    labels=['0', '1-2', '3-4', '5-7', '8+']
).astype('category')

# 5. Освобождаем память от исходных данных (если они больше не нужны)
del death_data

# 6. Инициализация KaplanMeierFitter
kmf = KaplanMeierFitter()

# 7. Построение кривых выживаемости с минимальным использованием памяти
plt.figure(figsize=(10, 6))

for group in analysis_data['kill_group'].cat.categories:
    mask = analysis_data['kill_group'] == group
    kmf.fit(
        durations=analysis_data.loc[mask, 'player_survive_time'].values,
        event_observed=analysis_data.loc[mask, 'event'].values,
        label=f'Убийств: {group}'
    )
    kmf.plot_survival_function(ci_show=False)

plt.title('Кривые выживаемости по группам убийств\n(Смерть = survive_time ≤ 1800 сек)')
plt.xlabel('Время в игре (секунды)')
plt.ylabel('Вероятность выживания')
plt.grid(True)
plt.axvline(x=1800, color='red', linestyle='--', alpha=0.3, label='Граница смерти (1800 сек)')
plt.legend()

# 8. Статистика по группам (вычисляем по мере необходимости)
print("\nСтатистика по группам:")
for group in analysis_data['kill_group'].cat.categories:
    mask = analysis_data['kill_group'] == group
    group_data = analysis_data.loc[mask, 'player_survive_time']
    print(f"\nГруппа {group}:")
    print(f"  Игроков: {mask.sum()}")
    print(f"  Среднее время: {group_data.mean():.1f} сек")
    print(f"  Медианное время: {group_data.median():.1f} сек")
    print(f"  Минимальное время: {group_data.min():.1f} сек")
    print(f"  Максимальное время: {group_data.max():.1f} сек")
    print(f"  Смертей: {analysis_data.loc[mask, 'event'].sum()}")

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KernelDensity
import time
from tqdm import tqdm  # для прогресс-бара

# Загрузка данных (вынесли в отдельную функцию для переиспользования)
def load_data():
    kills_data = pd.read_csv("./data/deaths/kill_match_stats_final_0.csv")
    kills_data['x_km'] = kills_data['victim_position_x'] / 100000  # см->км
    kills_data['y_km'] = kills_data['victim_position_y'] / 100000
    kills_data = kills_data[kills_data['map']=="ERANGEL"]
    kills_data = kills_data[kills_data['victim_position_x']!=0]
    print(kills_data.shape)
    return kills_data[['x_km', 'y_km']].dropna()

# Ускоренный KDE с настройкой параметров
def fast_kde(x, y, bandwidth=0.15, gridsize=100, sample_frac=None):
    if sample_frac is not None:
        idx = np.random.choice(len(x), int(len(x)*sample_frac), replace=False)
        x, y = x.iloc[idx], y.iloc[idx]
    
    kde = KernelDensity(bandwidth=bandwidth)
    data = np.vstack([x, y]).T
    kde.fit(data)
    
    grid = np.linspace(0, 8, gridsize)
    xx, yy = np.meshgrid(grid, grid)
    z = np.exp(kde.score_samples(np.vstack([xx.ravel(), yy.ravel()]).T))
    return xx, yy, z.reshape(xx.shape)

# Визуализация результатов
def plot_heatmap(xx, yy, z, title):
    plt.figure(figsize=(10, 8))
    plt.pcolormesh(xx, yy, z, cmap='hot', shading='auto')
    plt.colorbar(label='Плотность смертей')
    plt.title(title)
    plt.xlabel('X (км)')
    plt.ylabel('Y (км)')
    plt.show()

# Основной анализ
def analyze_hotspots(method='full', **kwargs):
    data = load_data()
    
    if method == 'full':
        print("Полный анализ всех данных...")
        xx, yy, z = fast_kde(data['x_km'], data['y_km'], **kwargs)
        plot_heatmap(xx, yy, z, "Полные данные (оптимизированный KDE)")
    
    elif method == 'sampled':
        print("Анализ с подвыборкой...")
        xx, yy, z = fast_kde(data['x_km'], data['y_km'], sample_frac=0.1, **kwargs)
        plot_heatmap(xx, yy, z, "10% случайных данных")
    
    elif method == 'fast':
        print("Ультра-быстрый анализ...")
        xx, yy, z = fast_kde(data['x_km'], data['y_km'],sample_frac=0.1, bandwidth=0.3, gridsize=50)
        plot_heatmap(xx, yy, z, "Упрощенный KDE (bandwidth=0.3, gridsize=50)")

# Сравнение методов
print("=== Сравнение методов расчета ===")
methods = {
    '1. Полный анализ': {'method': 'full'},
    '2. С подвыборкой 10%': {'method': 'sampled'},
    '3. Ультра-быстрый': {'method': 'fast'}
}


start = time.time()
analyze_hotspots('fast')
print(f"{time.time()-start:.2f} сек\n")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KernelDensity
import time
from PIL import Image  # для загрузки фонового изображения
from sklearn.utils import resample

def load_data(sample_frac=1.0):
    kills_data = pd.read_csv("./data/deaths/kill_match_stats_final_0.csv")
    kills_data['x_km'] = kills_data['victim_position_x'] / 100000  # см->км
    kills_data['y_km'] = kills_data['victim_position_y'] / 100000
    kills_data = kills_data[kills_data['map'] == "ERANGEL"]
    kills_data = kills_data[kills_data['victim_position_x'] != 0]
    
    if sample_frac < 1.0:
        kills_data = resample(kills_data, 
                           n_samples=int(len(kills_data)*sample_frac),
                           random_state=42)
    
    print(f"Используется {len(kills_data)} смертей (выборка {sample_frac*100}%)")
    return kills_data[['x_km', 'y_km']].dropna()

def fast_kde(x, y, bandwidth=0.15, gridsize=100):
    kde = KernelDensity(bandwidth=bandwidth)
    data = np.vstack([x, 8 - y]).T  # Отражаем по вертикали здесь
    kde.fit(data)
    
    grid = np.linspace(0, 8, gridsize)
    xx, yy = np.meshgrid(grid, grid)
    z = np.exp(kde.score_samples(np.vstack([xx.ravel(), yy.ravel()]).T))
    return xx, yy, z.reshape(xx.shape)

def plot_comparison(xx, yy, z):
    # Загрузка фонового изображения
    try:
        map_img = Image.open("data/erangel.jpg")
        map_img = np.array(map_img)
        extent = [0, 8, 0, 8]
    except FileNotFoundError:
        map_img = None
        print("Фоновое изображение не найдено, будет показана только тепловая карта")
    
    # Создаем фигуру с двумя субплогами
    plt.figure(figsize=(20, 8))
    
    # 1. Чистая тепловая карта
    plt.subplot(1, 2, 1)
    plt.pcolormesh(xx, yy, z, cmap='hot', shading='auto')
    plt.colorbar(label='Плотность смертей')
    plt.title('Тепловая карта смертей (отраженная)')
    plt.xlabel('X (км)')
    plt.ylabel('Y (км)')
    plt.gca().set_aspect('equal')
    
    # 2. Наложение на карту
    plt.subplot(1, 2, 2)
    if map_img is not None:
        plt.imshow(map_img, extent=extent, alpha=0.7)
    plt.pcolormesh(xx, yy, z, cmap='hot', shading='auto', alpha=0.5)
    plt.colorbar(label='Плотность смертей')
    plt.title('Наложение на карту Эрангеля')
    plt.xlabel('X (км)')
    plt.ylabel('Y (км)')
    plt.gca().set_aspect('equal')
    
    plt.tight_layout()
    plt.show()

def analyze_hotspots():
    data = load_data(0.1)
    
    start = time.time()
    xx, yy, z = fast_kde(data['x_km'], data['y_km'], 
                        bandwidth=0.2, gridsize=50)
    print(f"Расчет KDE занял {time.time()-start:.2f} сек")
    
    plot_comparison(xx, yy, z)

# Запуск анализа
analyze_hotspots()
```
